```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import xml.etree.ElementTree as ET
import re
```

```python
time = []
HR_rate = []
lat = []
lon = []
alt = []
dist = []
cadence = []
speed = []
watt = []
with open("C:/Users/cwh93/Downloads/exercise_tcx_file.tcx") as xml_file:
    xml_str = xml_file.read()
    xml_str = re.sub(' xmlns="[^"]+"', '', xml_str, count=1)
    root = ET.fromstring(xml_str)
    activities = root.findall('.//Activity')
    for activity in activities:
        tracking_points = activity.findall('.//Trackpoint')
        for tracking_point in list(tracking_points):
            children = list(tracking_point)
            time.append(children[0].text)
            lat.append(list(children[1])[0].text)
            lon.append(list(children[1])[1].text)
            alt.append(float(children[2].text))
            dist.append(float(children[3].text))
            HR_rate.append(int(list(children[4])[0].text))
#             HeartRateBpm.append(children[5].text)
#             speed.append(list(children[6])[0][0].text)
#             watt.append(list(children[6])[0][1].text)
            
```

```python
df = pd.DataFrame({'Time':time,
                      
                     'Latitude':lat,
                     'Logitude':lon,
                     'Altitude':alt,
                     'Distance_Meters':dist,
                  'HR Rate':HR_rate})
df
```

```python
df["geometry"]="POINT ("+df["Latitude"]+" "+df["Logitude"]+')'
```
