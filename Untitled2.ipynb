```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import xml.etree.ElementTree as ET
import re
import math
```

```python
time = []
HR_rate = []
lat = []
lon = []
alt = []
dist = []
cadence = []
speed = []
watt = []
with open("C:/Users/cwh93/Downloads/exercise_tcx_file.tcx") as xml_file:
    xml_str = xml_file.read()
    xml_str = re.sub(' xmlns="[^"]+"', '', xml_str, count=1)
    root = ET.fromstring(xml_str)
    activities = root.findall('.//Activity')
    for activity in activities:
        tracking_points = activity.findall('.//Trackpoint')
        for tracking_point in list(tracking_points):
            children = list(tracking_point)
            time.append(children[0].text)
            lat.append(list(children[1])[0].text)
            lon.append(list(children[1])[1].text)
            alt.append(float(children[2].text))
            dist.append(float(children[3].text))
            HR_rate.append(int(list(children[4])[0].text))
#             HeartRateBpm.append(children[5].text)
#             speed.append(list(children[6])[0][0].text)
#             watt.append(list(children[6])[0][1].text)
            
```

```python
df = pd.DataFrame({'Time':time,
                      
                     'Latitude':lat,
                     'Logitude':lon,
                     'Altitude':alt,
                     'Distance_Meters':dist,
                  'HR Rate':HR_rate})
df
```

```python
df["geometry"]="POINT ("+df["Latitude"].str[:-8]+" "+df["Logitude"].str[:-8]+')'
df
```

```python
for i in df.index:
    try:
        df.loc[i, "Direction"]=math.degrees(math.atan((float(df.loc[i+1,"Latitude"])-float(df.loc[i,"Latitude"]))/(float(df.loc[i+1,"Logitude"])-float(df.loc[i,"Logitude"]))))
    except:
        pass
```

```python
l_=list(df.index)
for i in df.index:
    try:
        if math.isnan(df.loc[i, "Direction"]) ==True:
            df.loc[i, "Direction"]=df.loc[i-1, "Direction"]
            
    except:
        pass
l_.reverse()
for l in l_:
    try:
        if math.isnan(df.loc[l, "Direction"]) ==True:
            df.loc[l, "Direction"]=df.loc[l+1, "Direction"]
            
    except:
        pass    
    
df
```

```python
list_index_dif=[]
for i in df.index:
    try:
        
        if df.loc[i,"geometry"]!=df.loc[i-1,"geometry"]:
            list_index_dif.append(i)
    except:
        pass
```

```python
df=df.loc[list(list_index_dif)]
df
```

```python
import folium
import numpy as np
import io
from PIL import Image, ImageDraw

for i in df.index:
    direct=df.loc[i,"Direction"]
    time=df.loc[i,"Time"]
    pushpin = folium.features.CustomIcon("C:/Users/cwh93/Downloads/wonmik.png", icon_size=(70,70))

    m = folium.Map(location=(44.8792964220047 , -122.64702451229095), zoom_start=15)
    folium.Marker(
        location=[44.8792964220047 , -122.64702451229095],
        icon=pushpin,
            height=1500,
        width=1500,
#          tooltip=str(direct)
    ).add_to(m)


    img_data = m._to_png(5)
    img = Image.open(io.BytesIO(img_data))
#     img=img.rotate(direct)

    width, height = img.size
    x = (width - height)//2
    img_cropped = img.crop((x, 0, x+height, height))

    # create grayscale image with white circle (255) on black background (0)
    mask = Image.new('L', img_cropped.size)
    mask_draw = ImageDraw.Draw(mask)
    width, height = img_cropped.size
    mask_draw.ellipse((0, 0, width, height), fill=255)

    # add mask as alpha channel
    img_cropped.putalpha(mask)





    img_cropped.save(str(i)+'.png')
    print('saved')

```
